### binary tree 이진트리
- 나무를 거꾸로 뒤집은 모양의 자료구조
- 이진트리가 구현이 쉽다
- 이진트리 왼쪽, 오른쪽 노드
- 이중 연결리스트 노드와 모양 동일

```python
class Tree:
    def __init__(self):
        self.nodes = []
        self.root = None

    def __str__(self):
        temp = []
        for node in self.nodes:
            temp.append(node.data)
        return f'{temp}'

    class TreeNode:        
        def __init__(self,data):
            self.data = data
            self.left = None
            self.right = None

    # ---------------
    # 트리에 노드 생성해서 자동으로 완전 이진 트리에 맞춰 추가
    # ---------------
    def append(self,data):
        node = self.TreeNode(data)
        self.nodes.append(node)

        if self.nodes.index(node) == 0:
            self.root = node; print('루트 노드 설정 실행완료')
        else:
            parent_index = (self.nodes.index(node)-1) // 2
            parent = self.nodes[parent_index]
            if parent.left == None:
                parent.left = node; print('왼쪽자식 추가 로직 실행')
            elif parent.right == None:
                parent.right = node; print('오른자식 추가 로직 실행')
            else: pass

    # ---------------
    # 전위 순회
    # ---------------
    def preorder(self):
        if self.root == None: return
        stack = [self.root]
        result = []

        while stack:
            node = stack.pop()
            result.append(node.data)
            if node.right:
                stack.append(node.right)
            if node.left:
                stack.append(node.left)

        print(" -> ".join(map(str,result)))


    # ---------------
    # 중위 순회
    # ---------------
    def inorder(self):
        if self.root is None : return

        stack = []
        node = self.root
        result = []

        while stack or node:
            while node:
                stack.append(node)
                node = node.left

            node = stack.pop()
            result.append(node.data)

            node = node.right

        print(' -> '.join(map(str,result)))
                

    # ---------------
    # 후위 순회
    # ---------------
    def postorder(self):
        if self.root is None : return

        stack = [self.root]
        result = []

        while stack:
            node = stack.pop()
            result.append(node.data)
            if node.left:
                stack.append(node.left)
            if node.right:
                stack.append(node.right)
            
        print(' -> '.join(map(str,result[::-1])))

            
```

```python
my_tree = Tree()
for i in range(1,8):
    my_tree.append(i)

print(my_tree)
my_tree.preorder()
my_tree.inorder()
my_tree.postorder()
```

#### 이진 트리 순회
- 노드를 한번씩 방문해서 전체를 다 도는 것 - 순회(Traversal)
    - 전위 순회(preorder) : "전"위, 루트를 제일 먼저,   루트 -> L -> R
    - 중위 순회(inorder) : "중"위, 루트를 중간에 ,      L -> 루트 -> R
    - 후위 순회(postorder) : "후"위, 루트를 마지막에,   L -> R -> 루트

#### 이진트리 사용 예
- 탐색, 정렬 - 이진 탐색트리
- 우선순위 큐 및 힙(OS, 메모리쪽)
- 다익스트라 최단거리 경로 알고리즘
- 수식 표현식 (사람이 보는 수식과 컴퓨터가 계산하는 수식은 다르다)
- 허프만 트리 (압축 알고리즘)
- DB 인덱싱
- 머신러닝, 의사결정트리
