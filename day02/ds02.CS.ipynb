### 32비트와 64비트

- 64 bit 와 32 bit의 주요 차이점

    - 메모리 주소 크기. 
        - 32비트 : $2^32$ (4GB) 까지의 메모리만 직접 주소 지정 가능
        - 64비트 : $2^64$ (16EB) 까지 메모리 주소지정 가능 (미친 증가량..)

    - CPU 레지스터 크기
        - 32비트 = 한번에 4바이트 연산가능
        - 64비트 = 한번에 8바이트 연산가능
            - 64비트에 레지스터 개수가 더 많다. but, 연산가능 크기 차이와 관계없이. 앞으로 더 많은 연산량을 요구하기에, 기대에 맞춰 개수가 늘어난 거라고 보자.
    
    - 리터럴 : 변수가 아니라 그 자체로 값을 의미 하는 것
        - 10, 300 , "문자열" ....

#### 배열의 동작방식

- 고정 크기 배열
    - 크기가 고정 되어있어, 크기보다 많은 데이터를 저장하려면 새로 배열을 만들어야한다.

- 동적 배열
    - 배열을 동적으로 할당, (힙 메모리에 할당) 프로그램 동작 중에 추가하려는 행동을 한다면, 더 큰 배열(보통 2배)을 만들고 거기에 기존 배열의 값을 복사 후 기존 배열을 해제한다.
    - 배열에 포인터가 포함되어있다면, 깊은복사
    - 보통의 배열은 얕은 복사
        - why? int, float 등은 매번 생성할 때 마다 새로운 메모리 공간을 할당 할 수도 있고, 레지스터 등에 저장 될 수도 있기 때문.
        - 즉, 기본 자료형의 데이터는 얕은 복사가 가능하다.
    - 기본 자료형은 어디에 저장되나?
        - 스택 메모리 : 함수 호출 시 지역변수 저장
        - 힙 메모리 : 동적 할당된 변수 저장
        - 데이터 영역 : 전역 변수, 정적 변수 저장
        - 코드 영역 : 상수 데이터가 저장 될 수도 있음.

#### 메모리의 영역

- 메모리 구조 [코드 데이터 스택 힙]
    - 이 메모리는 프로그램 하나 당 하나가 생성된다.
    - 그 크기는 OS가 판단하여 적절한 공간을 할당시켜준다.
    - 프로세스 ( 모두 독립적 ) , ex) 크롬브라우저 , vscode 동시 실행
    - 스레드( 스택만 공유 , 나머지는 개별) , ex) 크롬 브라우저의 여러 탭
    - 정적인 메모리 코드, 데이터는 컴파일시 크기가 결정되고
    - 동적인 메모리 중 스택은 자동으로 관리가 되며, 힙 메모리 영역은 필요할 때 마다 메모리에서 땡겨온다.

- 코드, 데이터, 스택, 힙 영역
    
    - 코드 영역 : 프로그램이 실행할 기계어코드(명령어) 가 저장되는 영역
    - 컴파일된 실행파일 (exe 등)의 코드가 올라감
    - 보통 ROM 메모리로 설정
        - 저장되는 데이터 : 리터럴, 함수코드 등

    - 데이터 영역 :  정적변수 Static, 전역변수 Global 데이터를 저장
        - 프로그램이 종료 될 때 까지 유지됨.
        - 초기화 여부에 따라 .data, .bss 로 나뉠 수 있다

    - 힙 영역 : 동적 할당 (new , malloc)된 데이터가 저장되는 영역
        - 크기가 고정되지 않고, 필요할 때 메모리를 할당 및 해제할 수 있다.
        - 프로그래머가 직접 관리 해야한다. (장점이자 단점)
        - 메모리 부족 시 **힙 오버플로우** 발생 가능.

    - 스택 영역 : 함수 호출 시 지역 변수와 매개변수(파라미터)가 저장됨
        - LIFO 방식 (후입선출) 로 동작
        - 함수가 끝나면, 자동으로 해제된다.
        - 재귀 호출 등을 남발하면 스택이 너무 많이 쌓여 **스택 오버플로우** 발생 위험이 있다.
