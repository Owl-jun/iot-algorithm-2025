### 1. P511 배낭함수 문제의 1, 2에 적합한 코드를 골라서 작성하시오. 

```python
maxWeight = 7
rowCount = 4
money = [0,13,8,6,12]
weight = [0,6,4,3,5]

def knapsack():
    array = [[0 for _ in range(maxWeight + 1)] for _ in range(rowCount + 1)]    # 1열, 1행은 더미 데이터를 넣기위해 maxWeight + 1, rowCount + 1 만큼 2차원 배열을 생성.
    for row in range(1, rowCount+1):                                            # 더미데이터를 제외한 1번부터 연산하기위해 start = 1, 같은이유로 rowCount+1
        for col in range(1, maxWeight+1):                                       # 위와 같음
            if weight[row] > col:                                               # 현재 선택된 원석의 무게가 가방에 들어가기에 더 무겁다면
                array[row][col] = array[row-1][col]                             # 현재 선택된 셀에 바로 윗칸의 데이터를 그대로 삽입하라.
            else :                                                              # 현재 선택된 원석의 무게가 가방에 들어가기에 알맞다면
                value1 = money[row] + array[row-1][col-weight[row]]             # 가방에 넣었으니, 원석 가치만큼 돈 + 여유 공간에 들어갈 원석이 있다면 그 원석의 가치 까지 합산
                value2 = array[row-1][col]                                      # 바로 윗칸에 삽입된 가격 가져오기 (현재 까지 같은 최대무게에서 최대 가격)
                array[row][col] = max(value1,value2)                            # 위 둘 중 더 큰값을 현재 셀에 저장합니다.
    return array[rowCount][maxWeight]                                           # 모든 연산이 끝나게되면, 마지막 행, 마지막 열에 가장 큰 가치의 결과가 나오게 됨으로 마지막셀을 리턴

knapsack()
```

### 2. P269 유명 맛집 대기줄 구현하기 문제를 while True:문을 사용하여 동적으로 입력을 받는 프로그램으로 변경작성하시오.

```python
class WaitingLine():                            # 대기줄 app 클래스 생성
    def __init__(self,size):                    # 생성과 동시에 이너클래스 원형queue 생성
        self.queue = self.circleQue(size+1)     # 사용자 경험을 위해 사이즈 크기 만큼 데이터를 넣기위해 +1 해줌 (원형큐라서)
    
    # 현재 대기줄상태 출력 메서드
    def state(self):
        print(f'현재 대기줄 상태 : {self.queue}')

    # 대기줄 인원 추가 메서드    
    def add_customer(self,data):
        self.queue.enQueue(data)    # enQueue 로 데이터삽입
        count = 0                   # 대기인원 담을 변수
        for _ in self.queue.que:    # 실제 큐 for문으로 순회
            if _ is not None:       # None 이아니라면 (데이터가 있다면)
                count += 1          # 인원수 +1
        print(f'{data} 님 대기 등록 완료, 현재 남은 대기인원 : {count} 명')
        self.state()                

    # 식당 입장 체크 메서드
    def welcome(self):
        temp = self.queue.deQueue()     # deQueue로 인원 삭제 후 리턴된 삭제된 데이터를 temp 변수로 받음
        print(f'{temp} 님 식당에 들어오셨습니다 ~!')    
        print('현재 대기줄 현황 : ',end='') 
        self.state()

    # 이너클래스 , 원형큐
    class circleQue():
        def __init__(self,size):    # 생성과 동시에 크기 만큼 기본값 None 으로 que 역할을 할 리스트 초기화
            self.size = size
            self.que = [None] * size
            self.start = 0          # 원형 큐 이기에 시작 인덱스 start, rear 모두 0
            self.rear = 0

        def __str__(self):          # 상태 출력 메서드
            _str = []
            for s in self.que:
                if s is not None:
                    _str.append(s)
            return "[" + ','.join(_str) + "]"
        
        def isFull(self):           # 큐가 꽉 찼는지 체크
            if (self.rear+1) % self.size == self.start:
                return True
            return False
        
        def isEmpty(self):          # 큐가 비었는지 체크
            if self.start == self.rear:
                return True
            return False
        
        def enQueue(self,data):     # 큐에 데이터 삽입
            if self.isFull():
                return
            self.rear = (self.rear + 1) % self.size
            self.que[self.rear] = data
        
        def deQueue(self):          # 큐 데이터 삭제 후 리스트나 스택의 pop처럼 삭제된 값 리턴
            if self.isEmpty():
                return
            self.start = (self.start + 1) % self.size
            temp = self.que[self.start]
            self.que[self.start] = None
            return temp
        


if __name__ == "__main__":          # 실제 앱 처럼 동작 시키고 싶었음..
    app = WaitingLine(5)
    while True:
        print()
        print('안녕하세요 ~! 기다리는 즐거움 캐치테이블 입니다!')
        print('성함을 입력해주세요 : [종료 : 0 , 입장 : 1]')
        print()
        name = input()
        if name == '0':
            print('대기할 인원이 없으므로 프로그램을 종료합니다.')
            break
        elif name == '1':
            if app.queue.isEmpty():
                print('입장할 인원이 없습니다.')
            else:
                app.welcome()
        else :
            if app.queue.isFull():
                print('대기 인원이 꽉찼습니다. 조금 후에 다시 이용해주세요. 앱을 종료합니다.')
                break
            else:
                app.add_customer(name)
```

### 3. P309 편의점 판매된 물건 목록 출력하기 문제 역시 while True: 문으로 동적으로 입력받는 프로그램으로 변경하시오.
데이터의 종류는 [ 콘말차, 삿포로 맥주, 아카페라 벤티 헤이즐넛, 레어치즈푸딩, 척오리지널 메가 사워, 요아정 요거트컵, 페퍼로니피자주먹밥, 널담 슈톨렌, 딸기마시멜로케이크, 버니공쥬 딸기뚱카롱, 고추잡채호빵, 체다 슈크림붕어스낵 ]

```python
import random   # 랜덤 함수 사용위해 임포트

class SellList:     # 셀리스트 클래스
    def __init__(self,dataAry=list,sellAry=list):   
        self.memo = []          # 메모리 역할
        self.dataAry = dataAry  # 입력받은 dataAry 를 클래스의 멤버변수로
        self.sellAry = sellAry  # 입력받은 sellAry 를 클래스의 멤버변수로
        self.root = None        # 루트노드는 None 으로 초기화
        self.append()           # SellList 의 append() 메서드 실행
        print('오늘 판매된 종류(중복X)--> ', end=' ')   # 출력 양식대로 출력
        self.preorder(self.root)                        # 트리 전위 순회
    class TreeNode:             # 실제 생성될 트리 노드 이너클래스
        def __init__(self, data):   # 데이터영역과 좌,우측 노드의 주소를 받을 포인터 변수를 가지고있음
            self.data = data
            self.left = None
            self.right = None

    def append(self):           # append 메서드 , 자동으로 리스트를 받아 부모설정후 이진탐색트리를 만들어준다.
        mother = self.TreeNode(self.sellAry[0])     # sellAry 의 첫번째 요소를 데이터로 트리노드생성
        self.root = mother                          # 첫번째 노드임으로 루트설정
        self.memo.append(self.root)                 # 메모리에 저장

        for name in self.sellAry[1:]:               # 첫번째 요소는 루트가 되었으니 [1:]1부터 슬라이싱
            node = self.TreeNode(name)              # 인덱스 1번요소 데이터 부터 끝까지 요소를 데이터로한 트리노드 생성
            current = self.root                     # 현재노드 선택 (루트)
            while True:                         
                if name == current.data:            # 중복방지코드
                    break
                if name < current.data:             # 네임이 루트보다 작다면 왼쪽 서브트리로 가렴
                    if current.left == None:
                        current.left = node
                        self.memo.append(node)
                        break
                    current = current.left
                else:                               # 루트보다 크다면 오른쪽 서브트리로 가렴
                    if current.right == None:
                        current.right = node
                        self.memo.append(node)
                        break
                    current = current.right
        print('이진 탐색 트리 구성 완료!')
    
    def preorder(self,node):                        # 전위 순회, 루트를 제일 먼저 탐색한다, Root -> Left -> Right 순회
        if node == None:
            return
        print(node.data, end= ' ')
        self.preorder(node.left)
        self.preorder(node.right)


    """
    입력 받을 데이터 목록
    '콘말차', '삿포로 맥주', '아카페라 벤티 헤이즐넛',
    '레어치즈푸딩', '척오리지널 메가사워', '요아정 요거트컵',
    '페퍼로니피자주먹밥','널담 슈톨렌','딸기마시멜로케이크',
    '버니공쥬 딸기뚱카롱', '고추잡채호빵', '체다 슈크림붕어스낵'
    """

if __name__ == "__main__":
    dataAry = []
    while True:
        command = input('판매중인 상품 명 입력 : [입력 종료 : 0 ]')
        if command == '0':
            break
        else:
            name = input()
            dataAry.append(name)


    sellAry = [random.choice(dataAry) for _ in range(20)]
    print('오늘 판매된 물건(중복O) --> ', end='')
    print('[' , ','.join(sellAry) , ']')
    app = SellList(dataAry,sellAry)
```

#### 4. P473 Self Study 13-2 의 문제. 이진 검색 알고리즘을 풀면서 검색 횟수도 같이 출력하도록 구현하세요.

```python
import random

def binSearch(ary,data):        # 이진 탐색트리
    pos = -1                    # pos = -1 뜻 : 아무것도 찾지 못한상태
    start = 0                   # ary의 시작 인덱스
    end = len(ary)-1            # ary의 마지막 인덱스
    count = 0                   # 검색횟수 카운트변수
    while start <= end:         # start 가 끝보다 커지면 검색실패 = 없다는 뜻 -1 리턴
        count += 1              # 검색횟수 증가
        mid = (start+end) // 2  # 중간값 찾기, 시작인덱스 + 끝인덱스 // 2
        if data == ary[mid]:    # 찾는값이 중간값과 같다면 , 찾은거니까 그대로 리턴
            return (mid,count)  # 
        elif data > ary[mid]:   # 데이터가 더 크다면 찾는 범위를 바꿔준다.
            start = mid+1       # mid+1 부터 end 까지, 범위가 절반으로 준다.
        elif data < ary[mid]:   # 위의 반대로 생각하면된다.
            end = mid-1
    return (pos,count)          # 못찾았을때 -1을 반환한다.

# _ary = random.sample([i for i in range(0,100001)],100000)     # 중복없이 10만개를 만들어보았으나, 이 테스트에선 별 의미없어보여 주석처리
_ary = [random.randint(0,100001) for _ in range(100000)]        # 중복있는 10만개의 랜덤한 숫자를 리스트에 담아보았다.
_ary.sort()                                                     # 이진탐색을 위해 정렬을 해주었다.
_data = random.randint(0,100001)                                # 찾을 값을 랜덤으로 뽑았다.
result = binSearch(_ary,_data)                                  # 결과를 튜플로 반환 받았다.

if result[0] == -1:                                             # pos = -1 이 반환되었다면, 없는거니까
    print(f'{_data} (이)가 없네요.')
else:                                                           # pos 값이 무언가 들어왔다면 위치를 찾은거니까
    print(f'{_data} (은)는 {result[0]} 위치에 있음')
print(f'## {result[1]}회 검색함')                               # 찾고 못찾고를 떠나, 일단 검색횟수 프린트
```
